```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy import sparse
from scipy.sparse.linalg import spsolve
from scipy.ndimage import gaussian_filter1d


dir = 'Raman/GNR_Jae Min_20251014'

def baseline_als(y, lam=1e6, p=0.01, niter=10):
    L = len(y)
    D = sparse.diags([1, -2, 1], [0, -1, -2], shape=(L, L-2))
    w = np.ones(L)
    for i in range(niter):
        W = sparse.diags(w, 0)
        Z = W + lam * D.dot(D.transpose())
        z = spsolve(Z, w*y)
        w = p * (y > z) + (1-p) * (y < z)
    return z

filename = 'cGnr_50x_0.1pc_40s_532 fluo_final.txt'

path = os.path.join(dir, filename)

# Read the data, skipping the header
data = np.loadtxt(path, skiprows=1)
with open(path, 'r') as f:
    header = f.readline().strip().split()

x = data[:, 0]
y = data[:, 1]

y_smooth = gaussian_filter1d(y, sigma=2)
baseline = baseline_als(y_smooth)
y_corrected = y - baseline
x = data[:, 0]
# Apply moving mean smoothing with window size 10
y = np.convolve(y_corrected, np.ones(5)/5, mode='same')


plt.plot(x, y)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (A.U.)')
plt.title('Coated AuNR Fluorescence Spectrum')
plt.grid(True)
plt.show()
```

```python
#filename = 'Ctab_50x_50pc_70s_1.txt'
filename = 'Ctab_50x_50pc_50s_1.txt'

path = os.path.join(dir, filename)

# Read the data, skipping the header
data = np.loadtxt(path, skiprows=1)
with open(path, 'r') as f:
    header = f.readline().strip().split()
x = data[:, 0]
y = data[:, 1]

baseline = baseline_als(y)
y_corrected = y - baseline
x = data[:, 0]
y = y_corrected

plt.plot(x, y)
plt.xlabel('Raman Shift (cm$^{-1}$)')
plt.ylabel('Intensity (A.U.)')
plt.title('CTAB Fluorescence Spectrum')
plt.grid(True)
plt.show()
```
